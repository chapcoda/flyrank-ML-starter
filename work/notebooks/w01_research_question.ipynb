## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**MY research question:**
Which content page should a review team open first for possible refresh, expansion, protection or monitoring - given they can only review a limited number each cycle?

**Unit of analysis**
One content page. One row = one page. The output ranks the pages.

**The dicision it improves**
The order in which a reviewer wprls through their limited capacity. Better ordering means a better line of review time on the pages most likely that needs attention/
**The action someone takes**
For an surfaced page, a reviewer decides to refresh, expand, protect, prune, or monitor it.

**Cost of wrong recommendation**
An false positive wastes a reviewers important time on a page that potentially did not need work. A false negative is worse, a genuinely declining, high traffic page stays unfixd and keeps losing traffic. Because the two errors cost differcently, choosing where to set the cutoff is an policy dicision.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

These three numbers show the problems in common, 54% decliing, the obious freshness rule would misfire, and enough pages have real exposure that misspriorizing them has a genuine traffic cost.


```python
# This cell is for CODE (numbers, a query, a check).

import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
else:
    # move up until we find the repo root (where data/raw lives)
    while not os.path.isdir("data/raw") and os.getcwd() != "/":
        os.chdir("..")

print("Working dir:", os.getcwd())
assert os.path.exists("data/raw/content_refresh_anonymized.csv"), "starter CSV not found"
print("Starter data found. You're ready.")

import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Number 1: how common is the problem? (declining rate)
declining_rate = (df["trend_direction"].str.lower() == "down").mean()
print(f"1. Declining rate: {declining_rate:.3f} "
      f"({(df['trend_direction'].str.lower() == 'down').sum():,} of {len(df):,} pages are trending down)")

# Number 2: does the 'stale = decline' assumption hold? median age by trend
age_by_trend = df.groupby("trend_direction")["content_age_days"].median().sort_values()
print("\n2. Median content age (days) by trend direction:")
print(age_by_trend.round(0).to_string())

# Number 3: is there enough visible demand to make review worthwhile?
visible = (df["impressions_90d"] >= 500).mean()
print(f"\n3. Share of pages with >=500 impressions in 90d: {visible:.3f} "
      f"(enough exposure that a wrong call actually costs traffic)")
```
